```python
import random
import csv

# Dictionaries to store original passwords and salted passwords
original_passwords = {}
salted_passwords = {}

class SaltingProjectFunctions:
    def __init__(self):
        pass
    # Function to salt a password
    def salt_password(self,password):
        # Add random elements to the password to make it more secure
        length = len(password)
        mid = length // 2
        salt = ''.join(random.choices('abcdefghijklmnopqrstuvwxyz1234567890', k=5))
        # Adding salt in the middle of the password
        salted_password = password[:mid] + salt + password[mid:]
        return salted_password

    # Function to write salted passwords to a file
    def write_passwords_to_file(self,filename):
        with open(filename, 'w', newline='') as file:
            writer = csv.writer(file)
            # Add headings to the CSV file
            writer.writerow(["username", "password", "salted_password"])
            for variable, password in original_passwords.items(): #variable is for username here
                salted_password = salted_passwords.get(variable, "")
                writer.writerow([variable, password, salted_password])

    # Function to read salted passwords from a file
    def read_passwords_from_file(self,filename):
        try:
            with open(filename, 'r') as file:
                reader = csv.DictReader(file)
                for row in reader:
                    variable = row["username"]
                    original_password = row["password"]
                    salted_password = row["salted_password"]
                    original_passwords[variable] = original_password
                    if salted_password:
                        salted_passwords[variable] = salted_password
        except FileNotFoundError:
            # Handle the case when the file doesn't exist
            print("File not found. No salted passwords loaded.")

    # Function to delete a password
    def delete_password(self,variable_name, filename):
        if variable_name in original_passwords:
            del original_passwords[variable_name]
        else:
            print("No such user exists")
        if variable_name in salted_passwords:
            del salted_passwords[variable_name]
            print(f"Password for {variable_name} deleted.")
        else:
            print("No such user exists in salted passwords dataset")
        obj1.write_passwords_to_file(filename)
#Creating object of the class
obj1=SaltingProjectFunctions()
while True:
    print("\nOptions:")
    print("1. Save a password \n2. Salt a password \n3. Show all passwords & Retrieve previously stored Data")
    print("4. Delete a password \n5. Exit\n")

    choice = input("Enter your choice (1/2/3/4/5): ")

    if choice == '1':
        print("\n")
        variable_name = input("Enter a Username: ")
        password = input("Enter a password: ")
        original_passwords[variable_name] = password
        do_salting = input("Do you want to add salt to this password? (yes/no): ")

        if do_salting.lower() == 'yes':
            salted_password = obj1.salt_password(password)
            salted_passwords[variable_name] = salted_password

        obj1.write_passwords_to_file('salted_passwords.csv')  # Store in a CSV file
        print("Password saved.")
        print()

    elif choice == '2':
        print("\n")
        variable_name = input("Enter the Username: ")
        if variable_name in original_passwords:
            password = original_passwords[variable_name]
            salted_password = obj1.salt_password(password)
            salted_passwords[variable_name] = salted_password
            obj1.write_passwords_to_file('salted_passwords.csv')  # Update the CSV file
            print("Password salted and updated.")
            print()
        else:
            print("Username not found.")
            print()

    elif choice == '3':
        print("\n")
        obj1.read_passwords_from_file('salted_passwords.csv')  # Read from the CSV file
        print("Stored Passwords:")
        for variable, original_password in original_passwords.items():
            print(f"Username: {variable}")
            print(f"Original Password: {original_password}")
            if variable in salted_passwords:
                salted_password = salted_passwords[variable]
                print(f"Salted Password: {salted_password}")
            else:
                print("No salted password.")
            print()
        print()

    elif choice == '4':
        print("\n")
        variable_name = input("Enter the Username to delete: ")
        obj1.delete_password(variable_name, 'salted_passwords.csv')
        #print(f"Password for {variable_name} deleted.")
        print()

    elif choice == '5':
        break
    else:
        print("Invalid option. Please select 1, 2, 3, 4, or 5.")
        print()
```
